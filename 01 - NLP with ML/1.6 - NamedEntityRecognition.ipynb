```python
corpus = "Virat Kohli is the best cricketer of the current generation who is inspiring a lot of young talent in the country"
```

```python
import nltk
```

```python
sentences = nltk.word_tokenize(corpus)
sentences
```

```python
tag_elements = nltk.pos_tag(sentences)
tag_elements
```

```python
nltk.download('maxent_ne_chunker_tab')
nltk.download("words")
```

```python
nltk.ne_chunk(tag_elements).draw()

# You will find a graph describing all the entities in the corpus.
```
