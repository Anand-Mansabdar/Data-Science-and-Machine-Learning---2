```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("goes", pos='v') # since 'goes' is a verb we have to give pos as 'v'

# Different pos values
# Noun - 'n' 
# Verb - 'v' 
# Adjective - 'a' 
# Adverb - 'r'
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programmed", "heats", "heated", "heating", "histories", "transforming", "transformed"]

for word in words:
  print(f"{word} ----> {lemmatizer.lemmatize(word, pos='v')}")
```

```python
lemmatizer.lemmatize("histories", pos='n') # Since, history is a noun we have to use pos='n'
```

```python
print(lemmatizer.lemmatize("goes", pos='v'))
print(lemmatizer.lemmatize("sportingly", pos='v'))
print(lemmatizer.lemmatize("fairly", pos='a'))
```
