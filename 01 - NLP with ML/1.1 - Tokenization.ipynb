```python
import nltk as nt
```

```python
corpus = """Hello! Good evening. My name is Anand Mansabdar.
I am a Computer Science 4th year student in Hyderabad, Telangana.
"""
print(corpus)
```

```python
# Tokens - from senetence to a paragraph
from nltk.tokenize import sent_tokenize

documents = sent_tokenize(corpus) 
```

```python
print(type(documents)) # list
documents
```

```python
for document in documents:
  print(document)
```

```python
# Tokenization - paragraph to words and sentence to words
from nltk.tokenize import word_tokenize

tokens = word_tokenize(corpus)
tokens # each word in the corpus is a seperate token (includes duplicates)
```

```python
for token in documents:
  print(word_tokenize(token))
```

```python
from nltk.tokenize import wordpunct_tokenize
print(wordpunct_tokenize(corpus)) # this function treats the punctuations and other garmmatical words as seperate tokens
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer() # the fullstop (.) will not be treated as a seperate token except the last full stop - to indicate the end of the sentence
tokenizer.tokenize(corpus)
```
