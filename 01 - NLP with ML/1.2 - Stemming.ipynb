```python
# Consider the following words

words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programmed", "heats", "heated", "heating", "history", "transforming", "transformed"]
```

```python
# 1 - PorterStemmer
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

for word in words:
  print(f"{word} ----> {stemmer.stem(word)}")
```

```python
# Disadvantages of PorterStemmer - cannot predict a few words correctly
print(stemmer.stem("history"))
print(stemmer.stem("congratulations"))
```

```python
# 2 - RegexpStemmer Class
from nltk.stem import RegexpStemmer
regexStem = RegexpStemmer("ing$|s$|e$|able$", min=4)
print(regexStem.stem("eating"))
print(regexStem.stem("plays"))
print(regexStem.stem("fundable"))
print(regexStem.stem("seas"))
print(regexStem.stem("partitioning"))
```

```python
# 3 - Snowball Stemmer
from nltk.stem import SnowballStemmer
snowStem = SnowballStemmer(language="english")

for word in words:
  print(f"{word} ----> {snowStem.stem(word)}")
```

```python
print(stemmer.stem("fairly"), stemmer.stem("sportingly"))
print(snowStem.stem("fairly"), snowStem.stem("sportingly"))
```

```python
# Disadvantages of snow ball stemmer
print(stemmer.stem("goes"))
print(snowStem.stem("goes"))
print(regexStem.stem("goes"))
```
